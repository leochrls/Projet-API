# Analyse préambule sur les datasets

Analyse sur la méteo

```python
#importation de matplotlib pour visualisation graphique
import matplotlib.pyplot as plt
```

```python
Meteo_Data['data']=pd.to_datetime(Meteo_Data['data'])

# Création du graphique pour les températures
plt.figure(figsize=(10, 6))
plt.plot(Meteo_Data['data'], Meteo_Data['temperature_moy'], color='tab:blue', label='Température Moyenne')

# Ajout des titres et labels
plt.title('Évolution des Températures Moyennes', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Température Moyenne (°C)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

# Affichage
plt.show()
```

Analyse du nombres de vélo

```python
Velo2023['Date'] = pd.to_datetime(Velo2023['Date'])
Velo2024['Date'] = pd.to_datetime(Velo2024['Date'])
```

```python
Velo2023['Année'] = Velo2023['Date'].dt.year
Velo2024['Année'] = Velo2024['Date'].dt.year
```

```python
total_2023=Velo2023['counts'].sum()
total_2024=Velo2024['counts'].sum()
```

```python
total = pd.DataFrame({'Year': ['2023', '2024'],'Total Counts': [total_2023, total_2024]})
```

```python
# Création de l'histogramme
plt.figure(figsize=(8, 6))
bars =plt.bar(total['Year'], total['Total Counts'], color=['tab:blue', 'tab:orange'])

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Position horizontale (centre de la barre)
        height,  # Position verticale (au-dessus de la barre)
        f'{int(height)}',  # Texte affiché (converti en entier si nécessaire)
        ha='center', va='bottom', fontsize=12, color='black'
    )
    
# Ajouter titres et labels
plt.title('Comparaison des Comptages de Vélos par Année', fontsize=14)
plt.xlabel('Année', fontsize=12)
plt.ylabel('Nombre Total de Vélos', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.ticklabel_format(style='plain', axis='y')



# Affichage
plt.show()
```
